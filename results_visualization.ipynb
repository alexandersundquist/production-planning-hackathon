```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Production data
hourly_production = [
    2, 2, 3, 3, 4, 2, 2, 3, 3, 2, 2, 2,
    2, 2, 4, 4, 2, 2, 2, 3, 3, 2, 3, 3
]
product_names = {0: "Product 0", 1: "Product 1", 2: "Product 2", 3: "Product 3", 4: "Product 4"}

# Color map for each product
colors = {
    0: "#d62728",  # red
    1: "#ff7f0e",  # orange
    2: "#2ca02c",  # green
    3: "#1f77b4",  # blue
    4: "#9467bd"   # purple
}

# --- Gantt chart: Hourly schedule ---
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8), gridspec_kw={'height_ratios': [2, 1]})
ax1.set_title("Hourly Production Schedule", fontsize=14)
for hour, product in enumerate(hourly_production):
    ax1.broken_barh([(hour, 1)], (product, 0.9), facecolors=colors[product])

ax1.set_yticks(list(product_names.keys()))
ax1.set_yticklabels([product_names[i] for i in product_names])
ax1.set_xticks(range(0, 24))
ax1.set_xlabel("Hour of Day")
ax1.set_ylabel("Product")

# Legend
legend_handles = [mpatches.Patch(color=colors[i], label=product_names[i]) for i in product_names]
ax1.legend(handles=legend_handles, bbox_to_anchor=(1.05, 1), loc='upper left')

# --- Bar chart: Total production per product ---
ax2.set_title("Total Units Produced Per Product", fontsize=14)
totals = {i: hourly_production.count(i) for i in product_names}
ax2.bar([product_names[i] for i in totals], [totals[i] for i in totals], color=[colors[i] for i in totals])
ax2.set_ylabel("Units Produced")
ax2.set_xlabel("Product")

plt.tight_layout()
plt.show()
```
